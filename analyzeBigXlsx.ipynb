```python
# Import packages
import glob
import json
import os
import time
import re
import pandas as pd
import requests
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")
import matplotlib.pyplot as plt
```

```python
wines = pd.read_excel("allWines.xlsx", index_col=None)
# Words we wish to drop from the Style
drop_list = [
    "Californian",
    "Oregon",
    "Napa Valley",
    "Washington State",
    "New York",
    "Virginia",
    "Texas",
    "Arizona",
    "Red"
]

wines['Style'] = wines['Style'].replace('|'.join(drop_list), '', regex=True) #Drop those words
wines['Style'] = wines['Style'].str.strip() # clean leading and trailing whitespace
wines['Style'] = wines['Style'].str.split().str.join(" ") # Collapse multiple spaces to one space
wines.drop_duplicates(inplace=True) # Drop duplicates
wines.dropna(inplace=True) # Drop rows with missing columns
```

```python
for X in wines["Style"].unique():
    plt.figure()
    sns.boxplot(x="Style", y="Rating", data=wines.loc[wines["Style"] == X]).set(
        title=X
    )
    sns.despine(offset=10, trim=True)
 
```
